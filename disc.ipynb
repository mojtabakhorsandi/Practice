```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('heartp.xlsx')
df
```

```python
del df['Unnamed: 0']
```

```python
age = df['age']

agecat=[]
 
for i in range(len(age)):
    if (age[i]>=25 and age[i]<=45):
        agecat.append('young')
        
    elif (age[i]> 45 and age[i]<= 65):
        agecat.append('middle')
        
    elif (age[i]> 65) :
        agecat.append('old')
    
agecat
        
```

```python
df['age']=agecat
```

```python
df.head()
```

```python
 from sklearn.preprocessing import KBinsDiscretizer
kbins = KBinsDiscretizer(n_bins=3 ,encode='ordinal',
                        strategy='kmeans')

df[['chol']]=kbins.fit_transform(df[['chol']])
```
